## Exercício 1: Verifique se um número é positivo, negativo ou zero.

```python
#Verifique o(s) número(s) apresentados;
#Verique se há sinal de negativo(-) ou não e não é zero (0);
#Se houver e não for zero (0), diga que é positivo;
#Se não, diga que é negativo;
#Se for zero diga que é zero.
#Se for zero diga que é zero.

num = int(input('Digite um número: '))
if num > 0:
  print('Positivo')
elif num < 0:
  print('Negativo')
else:
  print('Zero')
```

## Exercício 2: Peça dois números e exiba o maior.

```python
#Verifique os dois números apresentados;
#Calcule qual deles é o menor e qual é o maior;
#Apresente o maior no resultado.

num1 = int(input('Digite o primeiro número: '))
num2 = int(input('Digite o segundo número: '))

if num1 > num2:
  print('O número maior é ', num1)
elif num1 == num2:
  print('Os números são iguais.')
else:
  print('O número maior é ', num2)

```

```python
#Verifique o ano apresentado;
#Com base no histórico de anos bissextos (de 4 em 4 anos), calcule se este é um deles.

ano = int(input('Digite um ano com 4 digitos: '))
if (ano % 4 == 0 and ano % 100 != 0) or ano % 400 == 0:
  print('É bissexto')
else:
  print('Não é bissexto')
```

## Exercício 12: Crie um menu para converter temperaturas entre Celsius, Fahrenheit e Kelvin.

```python
def conversao(t,temp):
  return {
      'F': (temp *9/5) +32 ,
      'K': (temp + 273)

  }.get(t,'Conversão não encontrada')

t = input('Digite a escala de temperatura F/K: ')
temp = float(input('Digite o valor da temperatura: '))
print(conversao(t,temp))
```

## Exercício 13: Crie um programa que traduza números de 1 a 5 para palavras (1 = um, 2 = dois...).

```python
def traducao(num):
  return {
      '1': "um",
      '2': "dois",
      '3': "três",
      '4': "quatro",
      '5': "cinco"

  }.get(num, 'não encontrado')

num = input('Digite um número de 1 a 5: ')
print(traducao(num))
```

## Exercício 15: Implemente um programa que identifica o dia da semana baseado em um número (1 = Segunda, 7 = Domingo).

```python
def diasemana(num):
  return {
      '1': "segunda",
      '2': "terça",
      '3': "quarta",
      '4': "quinta",
      '5': "sexta",
      '6': "sábado",
      '7': "domingo"

  }.get(num, 'não encontrado')

num = input('Digite um número de 1 a 7: ')
print(diasemana(num))
```
